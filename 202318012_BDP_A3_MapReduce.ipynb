```python
# splitting words through flatMap()

rdd_split = rdd.flatMap(lambda x: x.split(' '))
words = rdd_split.collect()
words
```

```python
word_counts = rdd_split.map(lambda x: (x,1))
word_counts.collect()
```

```python
# reducing words i.e. assigning count of words by reducebykey()

rdd_count = word_counts.reduceByKey(lambda a,b: a+b)
rdd_count.collect()
```

```python
# sorting data 

rdd_sorted = rdd_count.map(lambda x: (x[1],x[0])).sortByKey()
rdd_sorted
```

```python
# printing first three records 

first_three = rdd_count.take(3)
for f in first_three:
    print(f" : {f[0]} Value: {f[1]} ")
```

```python

total_words = rdd_sorted.reduce(lambda a,b: (a[0]+b[0],a[1]))
print(f"Total Words : {total_words[0]} ")
```

```python
# mapreduce without using pyspark map and reduce functions

data = rdd.collect()
data
```

```python
# splitting and mappping words

import re
# print(data)
pattern = r'[\n\s!@#$%^&*\(\)\{\}\[\]~`/\?<>,\.]+'
words = re.split(pattern, str(data))
words
```

```python
# assigning counts to each word

word_count = list(map(lambda x: (x, 1), words))
word_count
```

```python
# counting frequency of each word

word_counts = {}
for word in words:
    word = word.lower()
    word_counts[word] = word_counts.get(word, 0) + 1

print(f'Total number of words:  {sum(word_counts.values())} ')
```

```python

print('\n\nList of Words: ')
print('\n\nWord \t\tFrequency ')
for key, values in word_counts.items():
    if key == '':   continue
    print(f' {key}: \t {values} ')
    

# word_counts.items()
```
